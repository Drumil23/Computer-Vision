```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
from skimage import data
from skimage.color import rgb2gray
```

```python
image = data.chelsea()
```

```python
image_gray = rgb2gray(image)
```

```python
image_b_low_1= cv2.GaussianBlur(image_gray,(31,31),1)
image_b_low_2= cv2.GaussianBlur(image_gray,(31,31),2)
image_b_low_3= cv2.GaussianBlur(image_gray,(31,31),3)
image_b_low_4= cv2.GaussianBlur(image_gray,(31,31),4)
image_b_low_5= cv2.GaussianBlur(image_gray,(31,31),5)
image_b_low_6= cv2.GaussianBlur(image_gray,(31,31),6)
image_b_low_6_5 = cv2.GaussianBlur(image_gray,(31,31),6.5)
```

```python
plt.imshow(image_b_low_1,cmap='gray')
```

```python
image_b_high= cv2.GaussianBlur(image_gray,(31,31),7)
```

```python
plt.imshow(image_b_high,cmap='gray')
```

```python
dog1=image_b_low_1-image_b_high
dog2=image_b_low_2-image_b_high
dog3=image_b_low_3-image_b_high
dog4=image_b_low_4-image_b_high
dog5=image_b_low_5-image_b_high
dog6=image_b_low_6-image_b_high
dog6_5=image_b_low_6_5-image_b_high
```

```python
fig, axs = plt.subplots(1, 7, figsize=(30, 30))

axs[0].imshow(dog1, cmap='gray')
axs[1].imshow(dog2, cmap='gray')
axs[2].imshow(dog3, cmap='gray')
axs[3].imshow(dog4, cmap='gray')
axs[4].imshow(dog5, cmap='gray')
axs[5].imshow(dog6, cmap='gray')
axs[6].imshow(dog6_5, cmap='gray')
```

```python
mx=np.max(dog1)
th=0.2*mx
dog1_th = np.where(dog1>th,255,dog1)
plt.imshow(dog1_th, cmap='gray')

dog2_th = np.where(dog2>th,255,dog2)
plt.imshow(dog2_th, cmap='gray')

dog3_th = np.where(dog3>th,255,dog3)
plt.imshow(dog3_th, cmap='gray')

dog4_th = np.where(dog4>th,255,dog4)
plt.imshow(dog4_th, cmap='gray')

dog5_th = np.where(dog5>th,255,dog5)
plt.imshow(dog5_th, cmap='gray')

dog6_th = np.where(dog6>th,255,dog6)
plt.imshow(dog6_th, cmap='gray')

dog6_5_th = np.where(dog6_5>th,255,dog6_5)
plt.imshow(dog6_5_th, cmap='gray')
```

```python
fig, axs = plt.subplots(1, 7, figsize=(30, 30))

axs[0].imshow(dog1_th, cmap='gray')
axs[1].imshow(dog2_th, cmap='gray')
axs[2].imshow(dog3_th, cmap='gray')
axs[3].imshow(dog4_th, cmap='gray')
axs[4].imshow(dog5_th, cmap='gray')
axs[5].imshow(dog6_th, cmap='gray')
axs[6].imshow(dog6_5_th, cmap='gray')
```

```python
image2 = data.rocket()
plt.imshow(image2)
```

```python
image2_gray = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
plt.imshow(image2_gray, cmap='gray')
```

```python
image2_b_low_1= cv2.GaussianBlur(image2_gray,(31,31),1)
image2_b_low_2= cv2.GaussianBlur(image2_gray,(31,31),2)
image2_b_low_3= cv2.GaussianBlur(image2_gray,(31,31),3)
image2_b_low_4= cv2.GaussianBlur(image2_gray,(31,31),4)
image2_b_low_5= cv2.GaussianBlur(image2_gray,(31,31),5)
image2_b_low_6= cv2.GaussianBlur(image2_gray,(31,31),6)
image2_b_high= cv2.GaussianBlur(image2_gray,(31,31),7)
```

```python
dog2_1 = image2_b_low_1-image2_b_high
dog2_2 = image2_b_low_2-image2_b_high
dog2_3 = image2_b_low_3-image2_b_high
dog2_4 = image2_b_low_4-image2_b_high
dog2_5 = image2_b_low_5-image2_b_high
dog2_6 = image2_b_low_6-image2_b_high
```

```python
mx=np.max(dog2_1)
th=0.2*mx
dog2_1_th = np.where(dog2_1>th,255,dog2_1)
dog2_2_th = np.where(dog2_2>th,255,dog2_2)
dog2_3_th = np.where(dog2_3>th,255,dog2_3)
dog2_4_th = np.where(dog2_4>th,255,dog2_4)
dog2_5_th = np.where(dog2_5>th,255,dog2_5)
dog2_6_th = np.where(dog2_6>th,255,dog2_6)
```

```python
fig, axs = plt.subplots(1, 4, figsize=(50, 50))

axs[0].imshow(dog2_1_th, cmap='gray')
axs[1].imshow(dog2_3_th, cmap='gray')
axs[2].imshow(dog2_5_th, cmap='gray')
axs[3].imshow(dog2_4_th, cmap='gray')
```
