```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data 
```

```python
image = cv2.imread('shapes.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
new_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(new_image)
```

```python
# Compute Harris corner response for new_image
new_img_harris = cv2.cornerHarris(new_image, 5, 5, 0.04)

# Get image size
n_rows, n_cols = new_img_harris.shape

# Set threshold for corner detection
n_th = 0.2 * new_img_harris.max()

# Loop through every pixel
for r in range(n_rows):
    for c in range(n_cols):
        if new_img_harris[r, c] > n_th:  # If it's a corner
            cv2.circle(new_image, (c, r), 5, (0, 255, 0), 1)  # Draw a circle

# Show the result
plt.imshow(new_image)
plt.show()
```
